```python
from app import *
import pandas as pd

history = ['','']
```

```python
# Инициализация истории вызовов функций
history = pd.DataFrame(columns=['function_name', 'function_result'])

# ID запроса для анализа
request_id = "744HDSD"

# Основной цикл работы агента
while True:
    # Формируем запрос с историей вызовов
    history_markdown = history.to_markdown() if not history.empty else "История пуста"
    user_message = f"""
История вызова функций:
{history_markdown}

ID: {request_id}"""
    
    answer = react_agent(user_message)
    print(f"Ответ агента: {answer}")

    next_function = answer['next_function']
    if next_function == "final_answer":
        break
    
    function_result = "Ошибок нет."
    new_row = {
        'function_name': next_function,
        'function_result': function_result
    }
    history = pd.concat(
        [history, pd.DataFrame([new_row])], 
        ignore_index=True
    )
    
# Выводим итоговую историю
print("\nИтоговая история вызовов:")
print(history.to_string(index=False))

```
